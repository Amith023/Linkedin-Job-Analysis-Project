```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Kept as a copy of uncleaned data.csv')
```

```python
def employee_count(inp):
    inp=str(inp)
    ret=''
    n=1
    for i in inp:
        if i=='-' and n==2:
            pass
        elif i =='-':
            n+=1
            ret+=i
        elif i in '0123456789-+':
            ret+=i
    if ret=='':
        ret=np.nan
    return ret

def applicants(inp):
    inp=str(inp)
    ret=''
    if inp=='' or inp=='Nil':
        ret=0
    else:
        for i in inp:
            if i in '0123456789':
                ret+=i
    if ret=='':
        ret=np.nan
    return ret
```

```python
df.Employees=df.Employees.apply(employee_count)
df.Applicants=df.Applicants.apply(applicants)
df.Followers=df.Followers.apply(applicants)
pd.set_option('display.max_rows', None)
```

```python
def job_level(inp):
    dot_position=inp.rfind('·')
    out=''
    if dot_position!=-1:
        out=inp[dot_position+2:]
    if len(out.strip())==0:
        out='N/A'
    return out
    


    
def job_type(inp):
    try:
        dot_position=inp.index('·')
        out=inp[:dot_position-1]
        out=out.strip()
    except:
        out=inp.strip()
    return out



df.Location=df.Location.apply(lambda x: x.lower())
certain_words=['north','east','south','west','contonment','urban','navi']
def only_city(inp):
    comma_count=inp.count(',')
    if comma_count==2:
        comma_postion=inp.find(',')
        city=inp[:comma_postion]
        city=city.strip().title()
        for word in certain_words:
            if word in city:
                city=city.replace(word,'')
                city=city.strip().title()

    else:
        city='N/A'
    return city


def state(inp):
    inp=inp.strip()
    comma_count=inp.count(',')
    if comma_count==2:
        out=inp.split(',')[1]
        out=out.strip()
        out=out.title()
    elif comma_count==1:
        out=inp.split(',')[0]
        out=out.strip()
        out=out.title()
    else:
        out='N/A'
    return out
```

```python
df['Job_type']=df.Job_level.apply(job_type)
df['Job_level']=df.Job_level.apply(job_level)
df['City']=df.Location.apply(only_city)
df['State']=df.Location.apply(state)
df['Country']='India'
```
